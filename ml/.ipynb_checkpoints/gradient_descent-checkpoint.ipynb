```python
# Implement Gradient Descent Algorithm to find the local minima of a function.
# For example, find the local minima of the function y=(x+3)Â² starting from the point x=2.
```

```python
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x + 3) ** 2

def df(x):
    return 2 * (x + 3)

x_old = 2
learning_rate = 0.1
precision = 0.0001
max_iterations = 1000

x_values = []
y_values = []

for i in range(max_iterations):
    gradient = df(x_old)
    x_new = x_old - learning_rate * gradient
    x_values.append(x_new)
    y_values.append(f(x_new))
    
    if abs(x_new - x_old) < precision:
        break
    
    x_old = x_new

print(f"Local minimum occurs at x = {x_new:.5f}")
print(f"Minimum value of function y = {f(x_new):.5f}")
```

```python
plt.plot(x_values, y_values, 'b-', label="Gradient Descent Path")
plt.title("Gradient Descent Convergence")
plt.xlabel("x")
plt.ylabel("y = (x + 3) ^ 2")
plt.legend()
plt.grid(True)
plt.show()
```
