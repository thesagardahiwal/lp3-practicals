```python
# Implement K-Nearest Neighbors algorithm on diabetes.csv dataset. Compute confusion
# matrix, accuracy, error rate, precision and recall on the given dataset.
# Dataset link : https://www.kaggle.com/datasets/abdallamahgoub/diabetes
```

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
df.head()
```

```python
x = df.drop("Outcome", axis = 1)
y = df["Outcome"]
```

```python
sns.countplot(x = y)
```

```python
y.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, random_state = 0, test_size = 0.25)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
from sklearn.metrics import classification_report
```

```python
y_pred = knn.predict(x_test)
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
error = []
for k in range(1, 41):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    error.append(np.mean(pred != y_test))
```

```python
error
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(16, 9))
plt.plot(range(1, 41), error, color='blue', marker='o', linestyle='dashed',
         markerfacecolor='red', markersize=8)

plt.title("K Value vs Error Rate")
plt.xlabel("Value of K")
plt.ylabel("Error Rate")
plt.grid(True)
plt.xticks(range(1, 41))
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 33)
```

```python
knn.fit(x_train, y_train)
```

```python
y_pred = knn.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```
